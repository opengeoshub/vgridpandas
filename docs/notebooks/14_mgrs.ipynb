### Latlon to MGRS

#### Pandas

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/opengeoshub/vopendata/main/csv/housing.csv')
resolution = 0
df = df.mgrs.latlon2mgrs(resolution, lat_col='lat', lon_col='lon')
df
```

#### GeoPandas

```python
import geopandas as gpd

df = gpd.read_file('https://raw.githubusercontent.com/opengeoshub/vopendata/main/shape/housing.geojson')
resolution = 0
df = df.mgrs.latlon2mgrs(resolution)
df
```

### MGRS to geo

```python
df = df.mgrs.mgrs2geo(mgrs_col='mgrs')
df.plot(edgecolor='white')
```

### MGRS binning

#### Pandas

```python
import pandas as pd

resolution = 1
df = pd.read_csv('https://raw.githubusercontent.com/opengeoshub/vopendata/main/csv/housing.csv')
stats = 'max'
numeric_col = 'median_house_value'
df_bin = df.mgrs.mgrsbin(resolution=resolution,
                    lat_col='lat',
                    lon_col='lon',
                    stats=stats,
                    numeric_col=numeric_col,
                    # category_col='ocean_proximity',
                    )
df_bin.plot(
    column=f'{numeric_col}_{stats}',
    cmap='Spectral_r',
    legend=True,
    linewidth=0.2,
)
```

#### GeoPandas

```python
import geopandas as gpd

resolution = 1
df = gpd.read_file('https://raw.githubusercontent.com/opengeoshub/vopendata/main/shape/housing.geojson')
stats = 'max'
numeric_col = 'median_house_value'
df_bin = df.mgrs.mgrsbin(resolution=resolution,
                    stats=stats,
                    numeric_col=numeric_col,
                    # category_col='ocean_proximity',
                    )
df_bin.plot(
    column=f'{numeric_col}_{stats}',
    cmap='Spectral_r',
    legend=True,
    linewidth=0.2,
)
```
